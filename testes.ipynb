```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D # Importar o toolkit para 3D

# 1. Definir os dados (exemplo)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y) # Criar uma malha para surface plots
Z = np.sin(np.sqrt(X**2 + Y**2)) # Função que define Z em relação a X e Y

# 2. Criar a figura e o eixo 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # Adicionar eixos 3D

# 3. Criar o gráfico (exemplo de surface plot)
surf = ax.plot_surface(X, Y, Z, cmap='viridis') # Mapear cores com um colormap

# 4. Adicionar rótulos aos eixos e um título
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
ax.set_title('Gráfico 3D de Superfície')

# 5. Adicionar uma barra de cores para o mapa de cores (opcional)
fig.colorbar(surf)

# 6. Exibir o gráfico
plt.show()
```
